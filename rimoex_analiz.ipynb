```python
def logisticpdf(x, u, s):
    a = math.e**-((x-u)/s)
    return a/(s*(1+a)**2)
def logisticcdf(x, u, s):
    a = math.e**-((x-u)/s)
    return 1/(1+a)
def raisedcospdf(x, u, s):
    return 1/(2*s) * (1 + np.cos((x-u)/s * math.pi))
def raisedcoscdf(x, u, s):
    a = (x-u)/s
    return (1+a+np.sin(a*math.pi)/math.pi)/2
```

## Строим графики

```python
e, d = ex, dx
norm_rv = sts.norm(e, d)
hs_rv = sts.hypsecant(e, 21.59)
laplace_rv = sts.laplace(e, 5)
logistic_rv = sts.logistic(0, 3.2)
```

```python
#генерируем функцию распределения для экспоненциального распределения
u, s = 0., 100
x = np.linspace(u-s,u+s,1000)
cdf = norm_rv.cdf(x)
lcdf = logistic_rv.cdf(x)
#lcdf = logisticcdf(x, 0.16, 3)
#ccdf = raisedcoscdf(x, u, s)
hscdf = hs_rv.cdf(x)
lacdf = laplace_rv.cdf(x)
plt.plot(x, cdf, label='norm CDF')
plt.plot(x, lcdf, label='logistic CDF')
#plt.plot(x, hscdf, label='hyperbolic secant CDF')
#plt.plot(x, lacdf, label='laplace CDF')

from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(data)
plt.step(ecdf.x, ecdf.y, label='ECDF')

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

```python
def pdfcombine(x, pis, mus, sigmas):
    return pis[0]*sts.norm(mus[0], sigmas[0]).pdf(x)+pis[1]*sts.norm(mus[1], sigmas[1]).pdf(x)
```

```python
npdf = norm_rv.pdf(x)
#lpdf = logisticpdf(x, 0, 2.9)
lpdf = logistic_rv.pdf(x)
hspdf = hs_rv.pdf(x)
lapdf = laplace_rv.pdf(x)
# строим график
plt.hist(data, bins=500, normed=True)
#plt.plot(x, lpdf, label='logistic PDF')
plt.plot(x, npdf, label='norm PDF')
#plt.plot(x, sts.norm(0, 6).pdf(x), label='asd')
#plt.plot(x, sts.norm(0, 30).pdf(x), label='asd')
pi = np.array([0.5, 0.5])
mu = np.array([0,0])
sigma = np.array([6,30])
plt.plot(x, pdfcombine(x, pi, mu, sigma), label='asd')
#plt.plot(x, hspdf, label='hyperbolic secant PDF')
#plt.plot(x, lapdf, label='laplace PDF')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

# # Случайная величина abs($\Delta$x)

найдем $\lambda ^{*}$ методом максимального правдоподобия $$\lambda ^{*}= \frac{n}{\sum_{i=1}^{n}x_{i}}$$

```python
lamb = np.sum(dataabs)/len(dataabs)
lamb
```

```python
halflogistic_rv = sts.halflogistic(0,3.1)
exp_rv = sts.expon(0,lamb)
```

```python
#генерируем функцию распределения для экспоненциального распределения
xx = np.linspace(0,120,1000)
hlcdf = halflogistic_rv.cdf(xx)
expcdf = exp_rv.cdf(xx)
plt.plot(xx, expcdf, label='exp CDF')
plt.plot(xx, hlcdf, label='half-logistic CDF')
#plt.plot(x, lcdf, label='logistic CDF')
#plt.plot(x, ccdf, label='raised cosine CDF')

from statsmodels.distributions.empirical_distribution import ECDF
ecdfabs = ECDF(dataabs)
plt.step(ecdfabs.x, ecdfabs.y, label='ECDF')

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

```python
exppdf = exp_rv.pdf(xx)
hlpdf = halflogistic_rv.pdf(xx)

# строим график
plt.hist(dataabs, bins=300, normed=True)
plt.plot(xx, exppdf, label='expon PDF')
plt.plot(xx, hlpdf, label='half-logistic PDF')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```
